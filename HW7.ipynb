#### Doppler shifts

Imagine an apparatus that shoots out Cs atoms at room temperature (roughly 1000 m/s). To slow the atoms, we have a laser focused at the opening of the aparatus which hits the Cs atoms with photons and excites the valence electron to its first excited state. Each photon from the laser exerts a directed exchange of momentum on the Cs atom. Meanwhile, when the excited Cs atom releases a photon, the photon trajectory is random. Hence the Cs atom experiences a net negative acceleration due to the laser. 

From the Cs atom's perspective, the laser's frequency is getting lower while in fact from the lab frame, the Cs is simply slowing down. As the Cs becomes almost stationary, the laser frequency is no longer high enough to excite the Cs atom. 

To compensate for the problem of not being able to excite the stationary Cs atom, we want to widen the gap between the ground state and excited state initially (to compensate for the shortening of the gap due to the Doppler shifting of the Cs atom) and then decrease our magnetic field relative to the speed decrease of our Cs atom with the goal being that the gap should remain constant over the course of the atom slowing down. In order to employ such a magnetic field we use a Zeeman slower (see diagram below).

```python
from IPython.display import Image

Image(filename='./Zeeman-B-field.jpg')
```

Since the widening of the energy gap between ground state and excited state from the Zeeman slower counteracts the shrinking of the gap from the Doppler shift, we have that the Doppler shift is equal to the negative Zeeman shift.

To calculate the Zeeman shift, we note that the Cs atom has no fine structure, therefore we have a strong-field Zeeman effect. Our Zeeman Hamiltonian is

$$
H'_Z = \frac{e}{2m}B_{ext}(L_z + 2S_z)
$$

For a Zeeman slower, the spacially inhomogenous magnetic field necessary to counteract the Doppler shift is

$$
B_{ext}(z)=B_{0}+B_{a}\sqrt{1-z/z_{0}}
$$
