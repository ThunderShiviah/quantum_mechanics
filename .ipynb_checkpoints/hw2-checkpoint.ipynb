#### **1.** (a) Why does it take 10 electrons to fill the $3d$ shell?

Since a d-shell implies $l = 2$, we have that there are a total of $m = 2l + 1$ different $d$ orbitals. Since we can fit two electrons per orbital (spin-up and spin-down), we can fit a total of $2*(2l + 1) = 2*(2*2 + 1) = 10$ electrons.

#### (b) Consider copper (element number 29). Explain why the ground state is designated $^2S_{1/2}$, Given that the ground state is Ar($4s$)($3d)^{10}$.

Since nickel is the element right before copper, let's look at nickel's configuration and see the different ways we could add an electron. 

Nickel has the configuration Ar($4s)^2$($3d)^{8}$. 

If we want to add another electron, seemingly the only place to add it would be to the ($3d)^{8}$ shell and raise it to ($3d)^{9}$, but then we would be left with an extremely electronegative $3d$ shell since it wants to complete to ($3d)^{10}$. Meanwhile the $4s$ shell is much more willing to give up electrons since it has $n = 4$ and experiences more screening than the $n = 3$ $d$ shell. As a result, the most stable configuration is for the $4s$ shell to transfer its electrons over to the very electronegative $3d$ shell and then pick up the extra electron leaving us with the Ar($4s$)($3d)^{10}$ configuration. 

We can verify that the ($4s$) orbital experiences more screening than the ($3d$) shell by looking at the effective nuclear charge of each in Copper. I found a table on wikipedia that provided the following data.

We can indeed see that for copper, the $4s$ orbital experiences much more shielding than the $3d$, as evidenced from the much lower effective nuclear charge (a paltry 5.842 to $3d$'s 13.201).

#### (c) Estimate the energy of the lonely $4s$ electron in copper, as follows: Ignore the 10 electrons in the $3d$ levels, and assume the Ar "core" electrons "screen" the charge of the nucleus to the maximum extent possible. What is the charge the $4s$ electron "sees" under this assumption? How far is the $4s$ electron removed from the nucleus under this assumption?

Since our electron is still contained within the atom, we know that the electron must feel a charge of at least $e$. Let's assume that our electron only feels a charge $e$ (since our core electrons are screening to the maximum extent possible). In that case, our electron effectively should behave as if it is the single electron in a hydrogen atom except that it still has $n = 4$ and $l = 0$ (the $s$ orbital).

Right off the bat we can get the energy of the electron from the Bohr Equation 

$$
E_n = -\left[\frac{m}{2\hbar^2}\left(\frac{e^2}{4\pi\epsilon_0}\right)^2\right]\frac{1}{n^2} = \frac{E_1}{n^2}
$$

Since n = 4, we have

$$
E_4 = \frac{E_1}{4^2} = \frac{-13.6 eV}{16} = -0.85 eV.
$$

Next, we can try to get the *most probable* radius of the electron by assuming the wavefunction to be that of Hydrogen and then solving for 

$$
\frac{dP}{dr} = 0
$$

where $P = |\psi|^2 4\pi r^2dr$ is the probability.

In this case, since for our electron $n = 4$ and $l = 0$, we have that 

$$
\psi_{40m} = R_{40}(r)Y^{m}_{0}(\theta,\phi).
$$

We can start by dividing out all terms that are constant with respect to $r$, namely $Y^{m}_{0}(\theta,\phi)$ and a few constant terms in $R_{40}(r)$ which I'll get to in a second as well as the $4\pi$ term that got tacked onto $|\psi|^2$.

Finally we are left with having to solve

$$
\frac{d}{dr}R_{40}(r)^2r^2 = 0
$$

Since I'm only making an approximation, I'll solve numerically.

```python
from sympy.physics import hydrogen, units
from sympy import *
init_printing()  #for pretty latex printing
```

First I declare my radius variable.

```python
## Note: everything is in Bohr radius units.
r = symbols('r', real = True, positive = True) # r should be non-negative, real number
```

Below I set up the calculation. I initially set $n = 1$ so I can confirm that I get back 1 (the units are Bohr radius).

```python
radial = hydrogen.R_nl(1,0,r)  # Returns the Hydrogen radial wavefunction.
radial
```

```python
f = diff((radial**2)*r**2, r) #the derivative of (square of the radial wavefunction multiplied by r^2).
f
```

```python
results = solve(f,r)
for result in results:
    print(N(result))
```

As expected, I got that the most probable radius of an electron in the ground state of hydrogen is the Bohr radius. Now we'll check the case where $n = 4$.

```python
radial = hydrogen.R_nl(4,0,r)  # Returns the Hydrogen radial wavefunction.
```

#### **2.** Consider two noninteracting particles with mass $m$ in a 1-D harmonic oscillator with frequency $\omega$. One particle is in the energy eigenstate $|n\rangle$, the other in  $|n+1\rangle$. Calculate the expectation value of the square of the distance between the two particles in the case that

#### (a) the particles are distinguishable.

#### (b) the particles are identical bosons with spin 0.

#### c) the particles are identical fermions with spin-1/2 in one of the triplet states.

**Hints**: In order to calculate $\langle x^2\rangle$, use the Virial Theorem, which in this particular case states that
$$
\langle T\rangle=\langle V\rangle
$$
in any stationary state of the harmonic oscillator. Use creation and annihilation operators to simplify the remaining calculations! See also Eq. [5.18] of the textbook and the discussion following it. 

### Solution

We want 
$$
\langle(x_{n+1} - x_n)^2\rangle = \langle x_{n+1}^2\rangle + \langle x_{n}^2\rangle - 2\langle x_{n+1}x_n\rangle.
$$

We have from chapter two that $\forall n,$

$$
\langle x \rangle = 0 \\
\langle x^2\rangle = (n + \frac{1}{2})\frac{\hbar}{m\omega}
$$

To get an initial answer, I'll assume that both particles are in the same spin state (so I can disregard spin).

**a.** From Eq. 5.19 we get that for distinguishable particles

$$
\langle(x_{n+1} - x_n)^2\rangle = \langle x^2\rangle_{n+1} + \langle x^2\rangle_{n} - 2\langle x\rangle_{n+1}\langle x\rangle_n
$$

Since $\langle x^2\rangle = (n + \frac{1}{2})\frac{\hbar}{m\omega}$ and $\langle x \rangle = 0$  $\forall n$, we have that

$$
\langle x_{n+1}^2\rangle + \langle x_{n}^2\rangle - 2\langle x_{n+1}x_n\rangle =
\left(n + \frac{3}{2}\right)\frac{\hbar}{m\omega} + \left(n + \frac{1}{2}\right)\frac{\hbar}{m\omega} + 0 \\
\implies \langle(x_{n+1} - x_n)^2\rangle = \left(n + 1\right)\frac{2\hbar}{m\omega}
$$

**b.** From Eq. 5.21 we get that for indistinguishable bosons

$$
\langle(x_{n+1} - x_n)^2\rangle = \langle x^2\rangle_{n+1} + \langle x^2\rangle_{n} - 2\langle x\rangle_{n+1}\langle x\rangle_n - 2|\langle x \rangle_{n,n+1}|^2
$$

where 

$$
|\langle x \rangle_{n,n+1}| \equiv \int x\psi_n(x)^*\psi_{n+1}(x)\mathrm{d}x.
$$

Since $\langle x^2\rangle = (n + \frac{1}{2})\frac{\hbar}{m\omega}$ and $\langle x \rangle = 0$  $\forall n$, we have that

$$
\langle x_{n+1}^2\rangle + \langle x_{n}^2\rangle - 2\langle x_{n+1}x_n\rangle - 2|\langle x \rangle_{n,n+1}|^2 =
\left(n + \frac{3}{2}\right)\frac{\hbar}{m\omega} + \left(n + \frac{1}{2}\right)\frac{\hbar}{m\omega} - 0 - 2|\langle x \rangle_{n,n+1}|^2.
$$

To evaluate the $2|\langle x \rangle_{n,n+1}|^2$ term, we refer to Eq. 3.98 which tells us

$$
\langle x \rangle_{n,n+1} = \langle \psi_n|x|\psi_{n+1}\rangle = \sqrt{\frac{\hbar}{2m\omega}}\left(\sqrt{n+1}\delta_{n,(n+1)-1} + \sqrt{n}\delta_{n+1,n-1}  \right) \\
\implies \langle x \rangle_{n,n+1} = \sqrt{\frac{\hbar (n + 1)}{2m\omega}}
$$

Hence

$$
\langle x_{n+1}^2\rangle + \langle x_{n}^2\rangle - 2\langle x_{n+1}x_n\rangle - 2|\langle x \rangle_{n,n+1}|^2 =
\left(n + \frac{3}{2}\right)\frac{\hbar}{m\omega} + \left(n + \frac{1}{2}\right)\frac{\hbar}{m\omega} - 0 - \frac{\hbar (n + 1)}{m\omega}. \\
\implies \langle(x_{n+1} - x_n)^2\rangle = (n + 1)\frac{\hbar}{m\omega}
$$

**c.** For indistinguishable fermions we again refer to Eq. 5.21, noting that for fermions we simply change the last sign on our equation from a minus to a plus:

$$
\langle(x_{n+1} - x_n)^2\rangle = \langle x^2\rangle_{n+1} + \langle x^2\rangle_{n} - 2\langle x\rangle_{n+1}\langle x\rangle_n + 2|\langle x \rangle_{n,n+1}|^2.
$$

Proceding as before, we find that 

$$
\langle x_{n+1}^2\rangle + \langle x_{n}^2\rangle - 2\langle x_{n+1}x_n\rangle - 2|\langle x \rangle_{n,n+1}|^2 =
\left(n + \frac{3}{2}\right)\frac{\hbar}{m\omega} + \left(n + \frac{1}{2}\right)\frac{\hbar}{m\omega} - 0 + \frac{\hbar (n + 1)}{m\omega}. \\
\implies \langle(x_{n+1} - x_n)^2\rangle = (n + 1)\frac{3\hbar}{m\omega}
$$

#### Regarding spin

Up until this point I have neglected spin. The above answers for bosons and fermions were dependent upon the *whole* configuration being symmetric or antisymmetric respectively. 

To take spin into account, first note that a spin zero state (for example a singlet) is anti-symmetric while a spin one state (a triplet) is symmetric. Hence, if we look again at part b. we see that while bosons are symmetric with respect to spacial coordinates, they are antisymmetric with respect to spin (spin zero) hence the whole system is anti-symmetric and the corrected answer would be what we found for part c (the anti-symmetric spacial system). Meanwhile the fermions in part c. are antisymmetric spacially but symmetric with respect to spin (since they are in the triplet configuration) hence overall they are antisymmetric. Thus the answer we found for part c. still applies to our fermion configuration.
